## Технологический процесс
Как золото получают из руды? Изучите внимательно стадии процесса.
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.




Опишем каждую стадию:
1. Флотация:
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка:
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

        calculation — расчётные характеристики

### Шаг 1.2. Проверить, что эффективность обогащения рассчитана правильно. Вычислить её на обучающей выборке для признака rougher.output.recovery. Найте MAE между произведенными расчётами и значением признака.

```python
# рассчитаем эффективность обогащения Recovery на train выборке
# формула Recovery = (C * (F - T)) / F * (C - T), где 
# F-rougher.input.feed_au
# С-rougher.output.concentrate_au
# T-rougher.output.tail_au
try: # ввиду того, что мы уже удалили столбцы из датафрейма df_train, откроем и заново загрузим исходный датафрейм 
    df_train_initial = pd.read_csv('e:/Яндекс.Практикум/Спринт 10 Сборный проект №2/data/gold_recovery_train_new.csv')
except FileNotFoundError:
    df_train_initial = pd.read_csv('/datasets/geo_data_0.csv')

rougher_output_recovery = df_train_initial['rougher.output.recovery'] 
C = df_train_initial['rougher.output.concentrate_au']
F = df_train_initial['rougher.input.feed_au']
T = df_train_initial['rougher.output.tail_au']
 
#рассчитаем Recovery (эффективность обогащения)
rougher_output_recovery_calc_train = (C*(F - T))/(F*(C -T)) * 100
 
#выведем на экран результат расчета, а также значение метрики МАЕ 
#print('Recovery (расчетная концентрация золота):', rougher_output_recovery_calc_train)
print('MAE (абсолютная средняя ошибка):', mean_absolute_error(rougher_output_recovery, rougher_output_recovery_calc_train))
```

```python
# дополнительно рассчитаем эффективность обогащения Recovery на full выборке
# формула Recovery = (C * (F - T)) / F * (C - T), где 
# F-rougher.input.feed_au
# С-rougher.output.concentrate_au
# T-rougher.output.tail_au
rougher_output_recovery = df_full['rougher.output.recovery'] 
C = df_full['rougher.output.concentrate_au']
F = df_full['rougher.input.feed_au']
T = df_full['rougher.output.tail_au']
 
#рассчитаем Recovery (эффективность обогащения)
rougher_output_recovery_calc_full = (C*(F - T))/(F*(C - T)) * 100
 
#выведем на экран результат расчета, а также значение метрики МАЕ 
#print('Recovery (расчетная концентрация золота):', rougher_output_recovery_calc_full)
print('MAE (абсолютная средняя ошибка):', mean_absolute_error(rougher_output_recovery, rougher_output_recovery_calc_full))
```

### Шаг 3.1. Написать функцию для вычисления итоговой sMAPE.

```python
def smape_function(target, predictions):
    smape = ((2 * abs(target - predictions) / (abs(target) + abs(predictions))).sum() * 100 / len(target))
    return smape
```

```python
def smape_total_function(target_test_ror, predictions_ror, target_test_for, predictions_for):
    ror_smape = smape_function(target_test_ror, predictions_ror)
    for_smape = smape_function(target_test_for, predictions_for)
    smape_total = ror_smape * 0.25 + for_smape * 0.75
    return smape_total

smape_score = make_scorer(smape_function, greater_is_better=False)
```

```python
# объявим функцию function_linear_regression_ds
def function_linear_regression_ds(features_train, target_train, features_test, target_test):
    model = LinearRegression(n_jobs=-1) # инициализирем модель LinearRegression
    model.fit(features_train, target_train) # обучим модель на тренировочной выборке
    predictions_test = model.predict(features_test) # получим предсказания модели на валидационной выборке
      
    result = (mean_squared_error(target_test, predictions_test))**0.5 # посчитаем значение метрики RMSE на валидационной выборке
    cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=RANDOM_STATE)
    scores_cross_validation = cross_val_score(model, features_test, target_test, scoring=smape_score, cv=cv)
           
    print("Cредний выход:", round(predictions_test.mean(), 1))
    print("RMSE (квадратный корень из средней квадратичной ошибки):", result)
    print("R2 (коэффициент детерминации) = ", model.score(features_test, target_test))
    print("MAE (среднее абсолютное отклонение)", mean_absolute_error(target_test, predictions_test))
    print('Значение smape:', scores_cross_validation)
    return predictions_test
```

####  Шаг 3.2.2. RandomForestRegressor()

```python
# объявим функцию random_forest_regressor_ds 
def random_forest_regressor_ds(features_train, target_train, features_test, target_test):
    best_model = None
    best_result = 10000
    best_est = 0
    best_depth = 0
    for est in range(10, 30, 10):
        for depth in range (1, 5):
            model = RandomForestRegressor(random_state=RANDOM_STATE, n_estimators=est, max_depth=depth, n_jobs=-1) # инициализируйте модель RandomForestRegressor с параметрами random_state=12345, n_estimators=est и max_depth=depth
            model.fit(features_train, target_train) # обучите модель на тренировочной выборке
            predictions_test = model.predict(features_test) # получите предсказания модели на валидационной выборке
            result = (mean_squared_error(target_test, predictions_test))**0.5 # посчитайте значение метрики rmse на валидационной выборке
            scores_cross_validation = cross_val_score(model, features_train, target_train, scoring = smape_score, cv = 5)
            result = smape_function(target_test, predictions_test)
            if result < best_result:
                best_model = model
                best_result = result
                best_est = est
                best_depth = depth
                best_predictions_test = predictions_test
    print("Cредний выход:", round(predictions_test.mean(), 1))
    print("RMSE (квадратный корень из средней квадратичной ошибки):", best_result, "Количество деревьев:", best_est, "Максимальная глубина:", depth)
    print("R2 (коэффициент детерминации) = ", model.score(features_test, target_test))
    print("MAE (среднее абсолютное отклонение)", mean_absolute_error(target_test, predictions_test))
    print('Значение smape:', scores_cross_validation)      
    return best_predictions_test  
```

    Разработанный прототип модели LinearRegression() немного уступает по оценке качества sMAPE другой модели прогнозирования по медианному значению (sMAPE = 9.05). Вариантом решения может стать более долгий поиск параметров для моделей "деревьев решений".
