# Определение перспективного тарифа для телеком-компании

## Описание проекта
Федеральный оператор сотовой связи "Мегалайн" предлагает клиентам два тарифных плана: «Смарт» и «Ультра». Коммерческий департамент хочет понять, какой тариф приносит больше денег, чтобы скорректировать рекламный бюджет.
Для достижения цели исследования предстоит сделать предварительный анализ тарифов на небольшой выборке клиентов - в распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Требуется проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

## Описание тарифов
### Тариф «Смарт»
    Ежемесячная плата: 550 рублей
    Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
    Стоимость услуг сверх тарифного пакета:
        минута разговора: 3 рубля
        сообщение: 3 рубля
        1 Гб интернет-трафика: 200 рублей

### Тариф «Ультра»
    Ежемесячная плата: 1950 рублей
    Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
    Стоимость услуг сверх тарифного пакета:
        минута разговора: 1 рубль
        сообщение: 1 рубль
        1 Гб интернет-трафика: 150 рублей
        
Примечание:
«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.
Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.

## Описание данных
    Таблица users (информация о пользователях):
        user_id — уникальный идентификатор пользователя
        first_name — имя пользователя
        last_name — фамилия пользователя
        age — возраст пользователя (годы)
        reg_date — дата подключения тарифа (день, месяц, год)
        churn_date — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
        city — город проживания пользователя
        tariff — название тарифного плана

    Таблица calls (информация о звонках):
        id — уникальный номер звонка
        call_date — дата звонка
        duration — длительность звонка в минутах
        user_id — идентификатор пользователя, сделавшего звонок

    Таблица messages (информация о сообщениях):
        id — уникальный номер сообщения
        message_date — дата сообщения
        user_id — идентификатор пользователя, отправившего сообщение
        
    Таблица internet (информация об интернет-сессиях):
        id — уникальный номер сессии
        mb_used — объём потраченного за сессию интернет-трафика (в мегабайтах)
        session_date — дата интернет-сессии
        user_id — идентификатор пользователя

    Таблица tariffs (информация о тарифах):
        tariff_name — название тарифа
        rub_monthly_fee — ежемесячная абонентская плата в рублях
        minutes_included — количество минут разговора в месяц, включённых в абонентскую плату
        messages_included — количество сообщений в месяц, включённых в абонентскую плату
        mb_per_month_included — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
        rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
        rub_per_message — стоимость отправки сообщения сверх тарифного пакета
        rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

```python
#посчитаем количество сделанных звонков и израсходованных минут разговора по месяцам для каждого пользователя

#округлим в таблице 'df_calls' длительность звонков до минут в большую сторону (это условие тарифа) 
#и сменим тип данных на int32 
df_calls["duration"] = np.ceil(df_calls["duration"])
df_calls["duration"] = df_calls["duration"].astype(np.int32)

#удостоверимся, что данные не более чем за один календарный год
print(df_calls["call_date"].min())
print(df_calls["call_date"].max())

#добавим столбец "month", в котором укажем номер месяца
df_calls["month"] = df_calls["call_date"].astype('datetime64[M]').dt.month

#сгруппируем таблицу по месяцам и "user_id", а также добавим сумму и переименуем столбцы для удобства
df_calls_month = df_calls.groupby(["month", 'user_id'])['duration'].agg(['sum', 'count']).reset_index().sort_values('user_id')
df_calls_month.columns = ["month", 'user_id', 'sum_duration', 'count_calls']

display(df_calls_month)
```

```python
#посчитаем количество отправленных сообщений по месяцам для каждого пользователя

#удостоверимся, что данные не более чем за один календарный год
print(df_messages["message_date"].min())
print(df_messages["message_date"].max())

#добавим столбец "month", в котором укажем номер месяца
df_messages["month"] = df_messages["message_date"].astype('datetime64[M]').dt.month

#сгруппируем таблицу по месяцам и "user_id", а также добавим сумму и переименуем столбцы для удобства
df_messages_month = df_messages.groupby(["month", 'user_id'])['message_date'].agg('count').reset_index().sort_values('user_id')
df_messages_month.columns = ["month", 'user_id', 'count_messages']

display(df_messages_month)
```

```python
#посчитаем объем израсходованного интернет-трафика по месяцам для каждого пользователя

#округлим в таблице 'df_calls' длительность звонков до минут в большую сторону (это условие тарифа) 
#и сменим тип данных на int32 

#удостоверимся, что данные не более чем за один календарный год
print(df_internet["session_date"].min())
print(df_internet["session_date"].max())

#добавим столбец "month", в котором укажем номер месяца
df_internet["month"] = df_internet["session_date"].astype('datetime64[M]').dt.month

#сгруппируем таблицу по месяцам и "user_id", а также добавим сумму и переименуем столбцы для удобства
df_internet_month = df_internet.groupby(["month", 'user_id'])['mb_used'].agg('sum').reset_index().sort_values('user_id')
df_internet_month.columns = ["month", 'user_id', 'mb_used_month']

#переведем Mb в Gb с учетом условия, что округление в большую сторону делается уже по итогам месяца
df_internet_month['gb_used_month'] = np.ceil(df_internet_month['mb_used_month'] / 1024)

display(df_internet_month)
```

```python
#объединим таблицу df_calls_month и df_messages_month
df_calls_messages = df_calls_month.merge(df_messages_month, on=['month', 'user_id'], how='outer')

#объединим таблицу df_internet_month и ранее объединенный датафрейм df_calls_messages (= df_messages_month + df_calls_month)
df_calls_messages_internet = df_calls_messages.merge(df_internet_month, on=['month', 'user_id'], how='outer')

#добавим к полученному сводному датафрейму df_calls_messages_internet данные по пользователям и тарифам
#сделаем объединение по how='right' датафрейма df_calls_messages_internet, чтобы учесть абонентов, у которых не было 
#ни звонков, ни смс, ни интернет активности
df_consolidated = df_calls_messages_internet.merge(df_users, on='user_id', how='right')

#проверим датафрейм df_consolidated на пропуски 
df_consolidated.info()
```

```python
#удалим "неактивных абонентов" (это разница между непустыми строками 'month' и непустыми 'user_id'), 
#т.к. их оказалось все 2 человека 
#"неактивных абонентов" можно было бы и оставить при значительном количестве, 
#но в таком случае не понятно, к какому месяцу относить их нулевые значения, 
#что будет портить картину при расчете средних значений
df_consolidated = df_consolidated.dropna(subset=['month'])

#заполним пропуски нулевыми значениями, что соответствует действительности (не было звонков - значит, 0 минут продолжительность звонков) 
df_consolidated = df_consolidated.fillna(0)

#приведем тип данных к логичному для каждого нужного нам столбца
df_consolidated["month"] = df_consolidated["month"].astype(np.int32)
df_consolidated["user_id"] = df_consolidated["user_id"].astype(np.int32)
df_consolidated["sum_duration"] = df_consolidated["sum_duration"].astype(np.int32)
df_consolidated["count_calls"] = df_consolidated["count_calls"].astype(np.int32)
df_consolidated["count_messages"] = df_consolidated["count_messages"].astype(np.int32)
df_consolidated["gb_used_month"] = df_consolidated["gb_used_month"].astype(np.int32)

print(df_consolidated.duplicated().sum())
df_consolidated.info()
```

```python
#переименуем столбец 'tariff_name' на 'tariff', чтобы можно было соединить таблицы
df_tariffs = df_tariffs.rename(columns={'tariff_name': 'tariff'})

#добавим к сводному датафрейму df_consolidated данные таблицы df_tariffs
df_consolidated = df_consolidated.merge(df_tariffs, on='tariff', how='outer')
display(df_consolidated)
df_consolidated.info()
```

```python
#создадим столбцы с разницей между фактическим использованием смс, звонков, интернета и входящей в тариф
df_consolidated['calls_excess'] = df_consolidated['sum_duration'] - df_consolidated['minutes_included']
df_consolidated['messages_excess'] = df_consolidated['count_messages'] - df_consolidated['messages_included']
df_consolidated['internet_excess'] = df_consolidated['gb_used_month'] - df_consolidated['mb_per_month_included'] / 1024

#создадим функцию, которая будет выводить 0, если не было перерасхода, и количество перерасхода, если таковой был 
def excess(tariff):
    if  tariff <= 0:
        return 0
    if  tariff > 0:
        return tariff

#применим функцию и создадим столбцы с перерасходом, если таковой был, и "0", если перерасхода не было 
df_consolidated['calls_excess_1_0'] = df_consolidated['calls_excess'].apply(excess)
df_consolidated['messages_excess_1_0'] = df_consolidated['messages_excess'].apply(excess) 
df_consolidated['internet_excess_1_0'] = df_consolidated['internet_excess'].apply(excess) 

display(df_consolidated)
```

```python
#добавим столбец с расчетом выручки по каждому клиенту
df_consolidated['profit'] = (df_consolidated['rub_monthly_fee'] 
                             + df_consolidated['calls_excess_1_0'] * df_consolidated['rub_per_minute']
                             + df_consolidated['messages_excess_1_0'] * df_consolidated['rub_per_message']
                             + df_consolidated['internet_excess_1_0'] * df_consolidated['rub_per_gb'])

display(df_consolidated)   
#df_consolidated.info()
```

## Шаг 4. Проверка гипотез
- [x] средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются
    <font color='orange'> <b>Гипотеза о равенстве выручки пользователей тарифов «Ультра» и «Смарт» отвергнута<b> </font>

<font color='orange'> <b>Гипотеза о равенстве выручки пользователей из Москвы и других регионав не может быть отвергнута<b> </font>
- [x] Пороговое значение alpha. <font color='orange'> <b>В качестве порогового значения выбрано значение 0,01 (для высокой точности)<b> </font>
- [x] Принципы формирования нулевой и альтернативной гипотез;
<font color='orange'> <b>Нулевая гипотеза H₀ всегда формулируется так, чтобы использовать знак равенства. Другими словами, чтобы все данные лежали в районе изначально предполагаемого значения.
Альтернативная гипотеза H₁ формируется как гипотеза, которая противоречит нулевой гипотезе: либо нулевая гипотеза не отвергается, либо отвергается и принимается H₁<b> </font>
- [x] Обоснование выбора критерия для проверки гипотез. <font color='orange'> <b>Критерий Стьюдента (t-тест) - это статистический метод, который позволяет сравнивать средние значения двух выборок и на основе результатов теста делать заключение о том, различаются ли они друг от друга статистически или нет. Причина использования обусловлена целью исследования - сравнение средних для определения более прибыльного тарифа. 
   Дополнительно: по ЦПТ можно сказать, что генеральная совокупность распределена нормально, поэтому t-test можно использовать в данном случае<b> </font>



### Шаг 4.1 Проверка гипотез по выручке тарифов «Ультра» и «Смарт»
Формулируем гипотезы:
   - [x] Гипотеза H0: средняя выручка пользователей тарифов «Ультра» и «Смарт» равны
   - [X] Гипотеза H1: средняя выручка пользователей тарифов «Ультра» и «Смарт» не равны

<font color='orange'> <b>Результат проверки гипотез: отвергаем гипотезу H0 (средняя выручка пользователей тарифов «Ультра» и «Смарт» равны)  (применяли t-тест)<b> </font>


```python
# проверяем гипотезу H0: средняя выручка пользователей тарифов «Ультра» и «Смарт» равны

#подготовим датафрейм по тарифу «Смарт»
df_profit_smart = df_consolidated.loc[df_consolidated['tariff'] == 'smart']
df_profit_smart = df_profit_smart.loc[:, ['month', 'city', 'profit']]  
#display(df_profit_smart.info())

#подготовим датафрейм по тарифу «Ультра»
df_profit_ultra = df_consolidated.loc[df_consolidated['tariff'] == 'ultra']
df_profit_ultra = df_profit_ultra.loc[:, ['month', 'city', 'profit']]  
#display(df_profit_ultra.info())

display("Статистики выручки по тарифу Ультра", df_profit_ultra['profit'].describe())
print('------------------------------------------')
display("Статистики выручки по тарифу Смарт",df_profit_smart['profit'].describe())
```

```python
# задаем критический уровень статистической значимости
alpha = 0.01 

# проверим гипотезу о равенстве средних двух независимых совокупностей
#задаем параметр equal_var = False, т.к. если выборки разного размера 
#и/или можно предположить, что дисперсии у генеральных совокупностей разные
results = st.ttest_ind(df_profit_ultra['profit'], df_profit_smart['profit'], equal_var = False)

# выведем на экран получившеся p-значение
print('p-значение:', results.pvalue) 

# сравним получившееся p-значение с заданным уровнем статистической значимости:
if results.pvalue < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Шаг 4.2 Проверка гипотез по выручке Москвы и других регионов
Формулируем гипотезы:
   - [x] Гипотеза H0: средняя выручка пользователей из Москвы равна выручке пользователей из других регионов
   - [x] Гипотеза H1: средняя выручка пользователей из Москвы не равна выручке пользователей из других регионов

<font color='orange'> <b>Результат проверки гипотез: гипотеза H0 (средняя выручка пользователей из Москвы равна выручке пользователей из других регионов) не может быть отвергнута (применяли t-тест)<b> </font>

```python
#print(df_consolidated['city'].unique())

# проверяем гипотезу H0: средняя выручка пользователей из Москвы равна выручке пользователей из других регионов

#подготовим датафрейм по Москве
df_profit_msk = df_consolidated.loc[df_consolidated['city'] == 'Москва']
df_profit_msk = df_profit_msk.loc[:, ['month', 'city', 'profit']]
#display(df_profit_msk)

#подготовим датафрейм по регионам
df_profit_regions = df_consolidated.loc[df_consolidated['city'] != 'Москва']
df_profit_regions = df_profit_regions.loc[:, ['month', 'city', 'profit']]
#display(df_profit_regions)

display(df_profit_msk['profit'].describe())
display(df_profit_regions['profit'].describe())
```

```python
# задаем критический уровень статистической значимости
alpha = 0.01 

# проверим гипотезу о равенстве средних двух независимых совокупностей
#задаем параметр equal_var = False, т.к. если выборки разного размера 
#и/или можно предположить, что дисперсии у генеральных совокупностей разные
results = st.ttest_ind(df_profit_msk['profit'], df_profit_regions['profit'], equal_var = False)

# выведем на экран получившеся p-значение
print('p-значение:', results.pvalue)

# сравним получившееся p-значение с заданным уровнем статистической значимости:
if results.pvalue < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
