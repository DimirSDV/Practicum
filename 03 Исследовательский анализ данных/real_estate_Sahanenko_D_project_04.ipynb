# Исследование объявлений о продаже квартир

Исходная база данных для исследования - это данные сервиса Яндекc Недвижимость (архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет). Требуется научиться определять рыночную стоимость объектов недвижимости, установив влияющие параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

##### Цель исследования - научиться определять рыночную стоимость объектов недвижимости в зависимости от параметров, которые указывает пользователь и в зависимости от параметров, которые могут быть получены автоматически из других источников и/или посчитаны

##### Для достижения цели исследования решим следующие задачи:
    1. Верхнеуровнево изучим датафрейм: общие сведения по датафрейму, уникальные значения
    2. Произведем предобработку данных: выявим и при необходимости удалим дубликаты, отработаем пропущенные значения, изменим типы данных
    3. Произведем добавление столбцов с расчетными параметрами, которые могут коррелировать с ценой
    4. Выявим параметры, влияющие на рыночную стоимость объектов недвижимости. Для этого: 
        4.1 изучим параметр площадь, цена, число комнат, высота потолков. Построим гистограммы
        4.2 изучим параметр время продажи для ответа на вопрос о средней скорости продажи. Построим гистограммы
        4.3 выявим и удалим редкие и выбивающиеся значения
        4.4 проведем анализ зависимости цены от площади, числа комнат, удалённости от центра, этажа, дня недели/месяца/года размещения
        4.5 изучим 10 населенных пунктов с наибольшим числом объявлений
        4.6 определим границу центра Санкт-Петербурга на основании цен
        4.7 изучим объявления о продаже квартир в центре Санкт-Петербурга

```python
#обработаем пропуски:

#1) заполняем пропуски в столбце "ceiling_height" медианным значением, 
#предварительно исправив отрицательные значения методом abs() на всякий случай
#2) замена медианным значением в данной выборке обоснована, т.к. высота потолков в большинстве 
#квартир приблизительно одинакова
#3) пропуски могли появиться ввиду того, что продавец не всегда знает высоту потолков, 
#например, выставляет объявление через мобильный телефон, находясь не в квартире, а также ввиду
#того, что данный параметр не сделан обязательным для заполнения. 
#Возможен также вариант с разноуровневыми потолками в многоэтажных квартирах или 
#разделение на 2 этажа (практикуется в СПб в исторической части, например, на В.О.)
df["ceiling_height"] = df["ceiling_height"].abs()
median_ceiling_height = df["ceiling_height"].median()
df["ceiling_height"] = df["ceiling_height"].fillna(median_ceiling_height)


#1) ввиду небольшого количества пропусков в столбце 'floors_total' (86 / 23699 = 0,36%) и невозможности заполнить пропуски
# конкретным значением (точный адрес не известен!) разумнее будет удалить данные строки. 
#Немаловажно, что данные в столбце количественные, поэтому, удалив дубликаты, мы сможем изменить тип данных в столбце на int 
#2) Причина появления пропусков - недостаточно жестко спроектированная база данных, в которой данный параметр не является обязательным
df = df.dropna(subset=["floors_total"]).reset_index(drop=True)

#1) заполняем пропуски в столбце "is_apartment" на False для удобства работы в дальнейшем. 
#2) замена на False в столбце логична, т.к. это категориальный тип данных, имеет всего 2 значения: либо апартаменты, либо нет
#3) пропуск появился, скорее всего, по причине архитектуры базы данных (параметр скорее всего выставляется "галочкой", а надо было делать выбор из списка!) 
df["is_apartment"] = df["is_apartment"].fillna(False)


#1) пропуски в столбце 'living_area' (1903 / 23699 = 8,0%) составляют значительную часть датафрейма, удалять их нельзя, 
#2) заполнить пропуски медианой или иным значением нельзя, т.к. жилую площадь разумнее рассчитать как разницу между общей площадью и площадью кухни,
#что не абсолютно верно (т.к. есть еще коридоры), но даст приближенное значение
#3) Причина появления пропусков - недостаточно жестко спроектированная база данных, в которой данный параметр не является обязательным 


#1) пропуски в столбце 'kitchen_area' (2278 / 23699 = 9,6%) составляют значительную часть датафрейма, удалять их нельзя, 
#2) заполнить пропуски медианой или иным значением нельзя, т.к. площадь кухни разумнее рассчитать как разницу между общей площадью и жилой площадью,
#что не абсолютно верно (т.к. есть еще коридоры), но даст приближенное значение
#3) Причина появления пропусков - недостаточно жестко спроектированная база данных, в которой данный параметр не является обязательным 
# Также возможны пропуски для квартир-студий (но это снова является недостатком базы данных, т.к. в таком случае должен проставляться 0)


#1-2) пропуски в столбце 'balcony' заполним 0, т.к. скорее всего продавцы не заполняли данный параметр именно при отсутствии балкона 
#(при наличии балкона большинство продавцов указали бы это как преимущество)
#3) Причина появления пропусков - недостаточно жестко спроектированная база данных, в которой данный параметр не является обязательным 
#(для исправления необходимо изначально оставлять поле пустым, затем выпающим списком предлагать выбор вариантов для продавца!)
df["balcony"] = df["balcony"].fillna(0)

#1-2) заменим пропуски в столбце 'locality_name' на "Не указан населенный пункт" (пропусков всего 0,2%, не влияют, но лучше не удалять из датафрейма)
#3) Причина появления пропусков - недостаточно жестко спроектированная база данных, в которой данный параметр не является обязательным 
#(для исправления необходимо изначально оставлять поле пустым, затем выпающим списком предлагать выбор вариантов для продавца, 
#а если нет подходящего варианта, необходимо указывать "Другой населенный пункт" !)
df["locality_name"] = df["locality_name"].fillna("Не указан населенный пункт")


#Пропуски в нижеперечисленных столбцах также не удаляем (ввиду значительного количества данных), но и заполнить медианой или иным значением нельзя.
# Часть параметров является субъективной оценкой
#(например, расстояние до прудов / водоемов, т.к. у людей могут различаться понятия).
# Вариантом заполнения пропусков может быть "-999", но в нашем случае оставим пропуски
#airports_nearest         5542
#cityCenters_nearest      5519
#parks_around3000         5518
#parks_nearest           15620
#ponds_around3000         5518
#ponds_nearest           14589

#1-2) заменим пропуски в столбце 'days_exposition' медианой 
#(количество пропусков составляет 13,4% датафрейма, пренебречь и удалить нельзя)
#Возможен вариант замены пропущенных значений на моду, но в данном случае остановимся на медиане
#3) Причина появления пропусков - вероятные ошибки в проектировании базы данных (данный столбец должен считаться автоматически)
df["days_exposition"] = df["days_exposition"].abs()
median_days_exposition = df["days_exposition"].median()
df["days_exposition"] = df["days_exposition"].fillna(median_days_exposition)

df.isna().sum()
```

### 3.1 Добавление в таблицу новых столбцов: цена квадратного метра, день недели, месяц и год публикации объявления, этаж квартиры (варианты — первый, последний, другой), соотношение жилой и общей площади, а также отношение площади кухни к общей

Произведено добавление в таблицу новых столбцов: цена квадратного метра, день недели, месяц и год публикации объявления, этаж квартиры (варианты — первый, последний, другой), соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
# добавляем в датафрейм столбец "цена квадратного метра"
df["price_m2"] = df["last_price"] / df["total_area"]

#приводим значения столбца 'first_day_exposition' в формат 'дата' 
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d %H:%M')

#добавляем в датафрейм столбец "день недели"
df["day_of_week"] = df['first_day_exposition'].dt.weekday

#добавляем в датафрейм столбец "месяц размещения объявления"
df["month_of_place_an_ad"] = df['first_day_exposition'].dt.month

#добавляем в датафрейм столбец "год размещения объявления"
df["year_of_place_an_ad"] = df['first_day_exposition'].dt.year

#создаем формулу для кластеризации по этажу "первый / другой / последний"
def floor_category(floor):
    if floor.loc['floor'] == 1:
        return 'первый'
    elif floor.loc['floor'] == floor.loc['floors_total']:
        return 'последний'
    else:
        return 'другой'

#добавляем в датафрейм столбец кластеризации по этажу "первый / другой / последний"
df['floor_category'] = df.apply(floor_category, axis=1)

#добавляем в датафрейм столбец "отношение жилой площади к общей"
df["share_living_to_total_area"] = df["living_area"] / df["total_area"]

#добавляем в датафрейм столбец "отношение площади кухни к общей"
df["share_kitchen_to_total_area"] = df["kitchen_area"] / df["total_area"]

#проверяем результат
display(df)
```

### 4.3 Удаление редких и выбивающихся значений и описание обнаруженных особенностей.

### 4.3 Вывод
При удалении редких и выбивающихся особенностей было обнаружено следующее: использовать стандартные статистические методы для определения границ выборки (1,5 межквартильного размаха) недопустимо по таким исследуемым параметрам, как 'total_area','days_exposition', 'last_price' и 'ceiling_height' (по всем, кроме 'rooms'). Поэтому увеличиваем границы интервалов для данных параметров до приемлемых (по смыслу и с учетом % удаляемых из анализа значений)

```python
#создадим функцию, которая будет строить график и очищать исходный датафрейм от значений за определенными значениями 
#(ориентируемся на межквартильный размах)
def filter_quantile_variable(data, column, low_iqr_cons, high_iqr_cons, plot):
    # computing Q1, Q3, IQR
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    # Filtering Values between (Q1 - low_iqr_cons * IQR) and (Q3 + high_iqr_cons * IQR)
    low_threshold = Q1 - (low_iqr_cons * IQR)
    high_threshold = Q3 + (high_iqr_cons * IQR)
    filtered_data = (data[(data[column] >= low_threshold) & (data[column] <= high_threshold)]).reset_index(drop=True)
         
    # Boxplot
    if plot: 
        filtered_data[column].plot(kind='box', grid=True)
        print(f"Процент данных с выбросами в {column}: {1 - filtered_data.shape[0] / data.shape[0]:.2%}")
        print(f"Нижний порог: {low_threshold:.2f}")
        print(f"Верхний порог: {high_threshold:.2f}")
        return filtered_data
    else:
        return filtered_data
```

```python
#удалим из df редкие и выбивающиеся значения по параметру 'total_area', применив функцию filter_quantile_variable
#в качестве параметров для определения границ при первой итерации используем 1,5 межквартильного размаха,
#для последующих итераций будем изменять значения параметров границ выборки 
#на основании общеизвестных представлений (жизненный опыт) и значения показателя "Процент данных с выбросами"
#проверим на разумность применения параметров (по показателям Текущее минимальное/максимальное значение), 
#оставим итоговые параметры и отфильтруем df
df = filter_quantile_variable(df, 'total_area', 1.5, 2.5, plot=True)
print(f"Текущее минимальное значение: {df['total_area'].min():.2f}")
print(f"Текущее максимальное значение: {df['total_area'].max():.2f}")
display(df)
```
