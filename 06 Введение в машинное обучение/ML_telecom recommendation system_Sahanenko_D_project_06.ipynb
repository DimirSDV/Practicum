### Шаг 2. Разделить исходные данные на обучающую, валидационную и тестовую выборки.

```python
# сразу добавим целевой признак (столбец) в датафрейм, приравняв 
df['ml_tariff'] = df['is_ultra']
df["ml_tariff"] = df["ml_tariff"].astype(np.int32)

# разделим сначала датафрейм на 2 части: 80% на обучающую + валидационную выборку и 20% на тестовую
df_train_valid, df_test = train_test_split(df, stratify=df['is_ultra'], test_size=0.20, random_state=12345)

# разделим полученный датафрейм df_train_valid на 2 части: 75% на обучающую выборку и 25% на валидационную
df_train, df_valid = train_test_split(df_train_valid, stratify=df_train_valid['is_ultra'], test_size=0.25, random_state=12345)

# посмотрим статистики по каждому датасету
display(df_train.describe())
display(df_valid.describe())
display(df_test.describe())
```

#### Вывод по "Шаг 2. Разделить исходные данные на обучающую, валидационную и тестовую выборки."
Исходные данные разделены на обучающую (60%), валидационную (20%) и тестовую выборки (20%).
По статистикам видно, что данные разделены с относительно приемлемыми отклонениями по медианным и средним значениям, но отклонения присутствуют (это особенно важно для проверки моделей на тестовых данных).

### Шаг 3. Исследовать качество разных моделей, меняя гиперпараметры. Кратко написать выводы исследования.

```python
#зададим целевой признак
df.loc[df['is_ultra'] == 0, 'ml_tariff'] = 0
df.loc[df['is_ultra'] == 1, 'ml_tariff'] = 1

features_train = df_train.drop(['is_ultra', 'ml_tariff'], axis=1)
target_train = df_train['ml_tariff']
features_valid = df_valid.drop(['is_ultra', 'ml_tariff'], axis=1)
target_valid = df_valid['ml_tariff']
features_test = df_test.drop(['is_ultra', 'ml_tariff'], axis=1)
target_test = df_test['ml_tariff']
```

```python
# DecisionTreeClassifier
best_model = None
best_result = 0
for depth in range(1, 50):
    model_dtc = DecisionTreeClassifier(random_state=12345, max_depth=depth) # обучим модель с заданной глубиной дерева
    model_dtc.fit(features_train, target_train) # обучим модель на тренировочной выборке
    predictions = model_dtc.predict(features_valid) # получим предсказания модели
    result = accuracy_score(target_valid, predictions) # посчитаем качество модели на валидационной выборке
    if result > best_result:
        best_model_dtc = model_dtc # сохраним наилучшую модель
        best_result = result #  сохраним наилучшее значение метрики accuracy на валидационных данных
        best_depth = depth
        
print("Accuracy наилучшей модели DecisionTreeClassifier на валидационной выборке:", best_result)
print("Depth (глубина 'дерева') = ", best_depth)
```

```python
# RandomForestClassifier
best_model = None
best_result = 0
for est in range(1, 70):
    model_rfc = RandomForestClassifier(random_state=12345, n_estimators=est) # обучим модель с заданным количеством деревьев
    model_rfc.fit(features_train, target_train) # обучим модель на тренировочной выборке
    predictions = model_rfc.predict(features_valid) # получим предсказания модели
    #predictions = model_rfc.predict_proba(features_valid)[:, 1]
    result = accuracy_score(target_valid, predictions) # посчитаем качество модели на валидационной выборке
    if result > best_result:
        best_model_rfc = model_rfc # сохраним наилучшую модель
        best_result = result #  сохраним наилучшее значение метрики accuracy на валидационных данных
        best_est = est
        
print("Accuracy наилучшей модели RandomForestClassifier на валидационной выборке:", best_result)
print("Est (количество 'деревьев') = ", best_est)
```

```python
# LogisticRegression
max_iter_not_auto = 300

model_lr = LogisticRegression(random_state=12345, max_iter=max_iter_not_auto, penalty='none') #установим значение количества итераций вручную
model_lr.fit(features_train, target_train)
model_lr.predict(features_train) 
model_lr.score(features_train, target_train)

dump(model_lr, 'model_9_1.joblib')
        
print("Accuracy модели LogisticRegression на валидационной выборке:", model_lr.score(features_train, target_train))
print("max_iter (максимальное количество итераций) = ", max_iter_not_auto)
```

#### Вывод по "Шаг 3. Исследовать качество разных моделей, меняя гиперпараметры. Кратко написать выводы исследования."

Было проведено обучение и оценка качества различных моделей со сменой гиперпараметров. По результатам оценки качества обучения модели расположились в следующем порядке:

    1) RandomForestClassifier. Accuraccy на валидационной выборке: 0.8164852255054432 при Est (количество 'деревьев') =  41
    2) DecisionTreeClassifier. Accuraccy на валидационной выборке: 0.8149300155520995 при Depth (глубина 'дерева') =  5
    3) LogisticRegression. Accuraccy на валидационной выборке:     0.7463692946058091 при max_iter (количество итераций) = 300



### Шаг 4. Проверить качество модели на тестовой выборке.

```python
# DecisionTreeClassifier проверка на тестовой выборке
predictions_dtc_test = best_model_dtc.predict(features_test) # получим предсказания лучшей модели dtc на тестовой выборке
result_dtc_test = accuracy_score(target_test, predictions_dtc_test) # посчитаем качество модели на тестовой выборке
        
print("Accuracy наилучшей модели DecisionTreeClassifier на тестовой выборке:", result_dtc_test)
#print("Depth (глубина 'дерева') = ", best_depth)
```
